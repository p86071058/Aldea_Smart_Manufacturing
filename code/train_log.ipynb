```python
def feature_extraction(data):
    uniques = pd.unique(data['P_ID'])
    concat = pd.DataFrame()
    for unique in uniques:
        target_pid = data.loc[data['P_ID'] == unique]
        target_pid.drop(['P_ID'], axis = 1, inplace = True)
        sort = target_pid.sort_values(by = ['TIME'])
        sort.drop(['TIME'], axis = 1, inplace = True)
        sort = encoder(sort)
        max_value = np.array(sort.max())
        min_value = np.array(sort.min())
        range_value = np.subtract(max_value, min_value)
        mean_value = np.array(sort.mean())
        std = np.array(sort.std())
        median = np.array(sort.median())
        varience = np.array(sort.var())
#         coe_var = np.divide(std, mean_value)
        ske = np.array(sort.skew())
        kur = np.array(sort.kurtosis())
#         length = np.array(len(sort))
        features = np.concatenate((max_value, min_value, range_value, mean_value, std, median, varience, coe_var,
                                   ske, kur))
        features = np.reshape(features, (1, 1450))
        features_df = pd.DataFrame(features)
        concat = pd.concat([concat, features_df], axis = 0)
    return concat
```

```python
def impute_encoder(data):
    for col in data.columns:
        unique = pd.unique(data[col])
        if len(unique) < 100:
            data[col].fillna(value = data[col].value_counts().index[0], inplace = True)
            fe = data.groupby(col).size() / len(data)
            data.loc[:, col + '_encode'] = data[col].map(fe)
            data.drop([col], axis = 1, inplace = True)
        else:
            data[col].fillna(value = data[col].mean(), inplace = True)
    return data
```
